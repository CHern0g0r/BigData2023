```python
import pandas as pd
import numpy as np

from pathlib import Path
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data_path = '../data'
lter_path = f'{data_path}/penguins_lter.csv'
size_path = f'{data_path}/penguins_size.csv'
print(Path(lter_path).resolve())
```

```python
ldf = pd.read_csv(lter_path)
print(len(ldf))
ldf.head(2)
```

```python
sdf = pd.read_csv(size_path)
print(len(sdf))
sdf.head(2)
```

```python
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15, 5))

v, c = np.unique(sdf['species'], return_counts=True)
print(v, c)
ax0.pie(c, labels=v)
ax0.set_title('species')
v, c = np.unique(sdf['island'], return_counts=True)
print(v, c)
ax1.pie(c, labels=v)
ax1.set_title('island')
v, c = ['male', 'female'], [np.sum(sdf['sex'] == 'MALE'), np.sum(sdf['sex'] == 'FEMALE')]
print(v, c)
ax2.pie(c, labels=v)
ax2.set_title('sex')
pass
```

```python
lengths = sdf[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm']].dropna()
mass = sdf['body_mass_g'].dropna()

f, (ax, ax1) = plt.subplots(1, 2, figsize=(10, 5))

ax.boxplot(lengths)
ax.set_xticklabels(
    lengths.columns,
    fontsize=8
)

ax1.boxplot(mass)
ax1.set_xticklabels(
    ['body_mass_g'],
    fontsize=8
)
pass
```
